### Line Plot:

Joining the group of points using a line

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y) ### plotting the graph for variables x and y
plt.show()
```

```python
plt.plot(x,y,'y')
plt.show()
```

```python
plt.plot(x,y,'g')
plt.title("Sample line plot")  #### title for the graph
plt.xlabel("x-axis")   ### name of the x-axis
plt.ylabel("y-axis")   ### name of the y-aixs
plt.show()
```

```python
import numpy as np
x = np.linspace(0,4,50)
x
```

```python
plt.plot(x,x,color = 'g',label = "sample-1")
plt.plot(x,x**2,color = 'm',label = "x^2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("line plot with two graphs")
plt.legend() ## visuilizing the names of the graph line in the plot.
plt.show()
```

### Subplots:

* subplot is a function used to create multiple plots at a time.
* subplot(number_of rows,number_of_columns)


```python
### creating subplots:

plt.subplots(2,2)   ## 2 rows and 2 columns

plt.show()
```

```python
fig,a = plt.subplots(2,2)
x = np.arange(1,5)  # [1,2,3,4]
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exponential')
a[1][1].plot(x,np.log2(x))
a[1][1].set_title('logaritham')
plt.tight_layout()
plt.show()
```

```python
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2),rowspan= 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2,colspan = 2)

x = np.arange(1,10)
a2.plot(x,x**2)
a2.grid(True)
a2.set_title("square")
a2.set_xlabel("x-axis")
a2.set_ylabel("y-axis")
#a1.plot(x,np.exp(x))
#a1.set_title("exponential")
#a1.grid(color = 'b',ls = '-.')
a3.plot(x,np.log(x))
a3.set_title("log")
plt.tight_layout()
plt.show()
```
