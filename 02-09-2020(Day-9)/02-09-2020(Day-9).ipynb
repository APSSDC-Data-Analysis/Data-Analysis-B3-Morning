```python
### Joint plot:  combination of two plots.

sns.jointplot(x = "sepal_length",y = "sepal_width",data = iris,kind = "reg")
plt.show()
```

```python
sns.jointplot(x = "petal_length",y = "petal_width",data = iris,kind = "reg")
plt.show()
```

```python
sns.jointplot(x ="petal_length",y = "petal_width",data = iris)
```

```python
sns.jointplot(x = "petal_length",y = "sepal_length",data = iris,kind = "kde")
plt.show()
```

```python
sns.jointplot(x = "petal_length",y = "sepal_length",data= iris,color = "k").plot_joint(sns.kdeplot,n_levels = 10)
```

```python
help(sns.jointplot)
```

### pairplot:

it is used to plot mutliple plots at a time for the entire data.

```python
sns.set_style("ticks")
sns.pairplot(iris,hue = "species",diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()



### hue : this selects the categorical data from the defiend column name.
```

```python
sns.pairplot(iris,hue = "species",markers = ["o","d","s"])
plt.savefig("seaborn.png")
plt.show()

```

### Stripplot:

```python
sns.stripplot(x= "species",y = "sepal_length",data = iris)
plt.show()
```

### Heatmap:


```python
a = np.random.randn(10,20)
g = sns.heatmap(a)
plt.show()
```

```python
sns.heatmap(a,vmin = 0,vmax = -5)
```

```python
plt.figure(figsize =(10,8))
sns.heatmap(a,annot = True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(a,annot = True,linewidth = 0.5)
```

```python
sns.heatmap(a,cmap = "YlGnBu")
```
