## Scaling Techniques
- Standard Scaler
- robust Scaling
- Normalization
- MinMax Scaler

### Standard Scaler
- Standard scaler assumes your data is normally distributed and apply feature scaling on data
    - Z = [Xi - mean(X)] / std(X)

```python
pip install sklearn
```

```python
from sklearn import preprocessing
```

```python
scl1 = preprocessing.StandardScaler()
```

```python
scl1
```

```python
scl_data = scl1.fit_transform(data)
scl_data
```

```python
data.columns
```

```python
scl_data = pd.DataFrame(scl_data, columns = data.columns)
scl_data
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))
ax1.set_title('Without Scaling')
sns.kdeplot(data['TV'],ax = ax1)
sns.kdeplot(data['radio'],ax = ax1)
sns.kdeplot(data['newspaper'],ax = ax1)
sns.kdeplot(data['sales'],ax = ax1)

ax2.set_title('After Applying Scaler')
sns.kdeplot(scl_data['TV'],ax = ax2)
sns.kdeplot(scl_data['radio'],ax = ax2)
sns.kdeplot(scl_data['newspaper'],ax = ax2)
sns.kdeplot(scl_data['sales'],ax = ax2)
```

## Robust Scaling
- [Xi - Q1(X)] / [Q3(X) - Q1(X)]

```python
data.describe()
```

```python
scl_d2 = preprocessing.RobustScaler()
scl_d2
```

```python
scl_data = scl_d2.fit_transform(data)
scl_data
```
