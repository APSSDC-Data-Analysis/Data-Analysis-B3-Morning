### Working with missing value
- 4 main methods to identify and manuplate missing values
    - isnull()
    - notnull()
    - dropna()
    - fillna()
    
- **Note : Always have a backup of original data**

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('market_fact.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

- dropna(): Delete the missing values

```python
d = data.dropna(axis = 0)
```

```python
d
```

```python
d.shape
```

```python
d.isnull().sum()
```

```python
data.info()
```

- fillna()
    - Replace the null values with either 0 or mean(), if data is numeric
    - Replace the null values with any other object, if data is categorical

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data2 = data.copy()
```

```python
data2.head()
```

```python
data2.fillna(0, inplace = True)
```

```python
data2
```

```python
data2.shape
```

```python
data2.isnull().sum()
```

```python
data3 = data.copy()
```

```python
data3.head()
```

```python
data3.shape
```

```python
data3.isnull().sum()
```

```python
data3['Product_Base_Margin'].mean()
```

```python
data3.describe()
```

```python
data3['Product_Base_Margin'] = data3['Product_Base_Margin'].fillna(data3['Product_Base_Margin'].mean())
```

```python
data3
```

```python
data3.shape
```

```python
data3.isnull().sum()
```
