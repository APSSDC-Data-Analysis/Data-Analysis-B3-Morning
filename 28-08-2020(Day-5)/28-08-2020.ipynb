### Combining

```python
dict1 = {'Name':['lavanya','niharika','nagamounika'],
        'Branch':['Cse','Civil','mech']}
d1 = pd.DataFrame(dict1)
d1
```

```python
dict2 = {'Name':['alekhya','girrisha','naga'],
        'Branch':['ece','it','eee']}
d2 = pd.DataFrame(dict2)
d2
```

```python
pd.concat([d1,d2])
```

- axis = 0 --> rows
- axis = 1 --> columns

```python
pd.concat([d1,d2],axis = 1)
```
